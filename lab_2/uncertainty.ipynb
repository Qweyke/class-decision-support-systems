```python
import pandas as pd
import numpy as np

price = 210 
cost = 90  

demand_values = np.array([8, 9, 10, 11, 12])
days_count = np.array([15, 25, 30, 20, 10])
alternatives = demand_values.copy()

profit_matrix = np.zeros((len(alternatives), len(demand_values)))

for i, x in enumerate(alternatives):
    for j, y in enumerate(demand_values):
        sold = min(x, y)
        revenue = sold * price
        total_cost = x * cost
        profit_matrix[i, j] = revenue - total_cost

df_profit = pd.DataFrame(
    profit_matrix,
    index=alternatives,
    columns= demand_values,
)
df_profit.index.name = "Alt-s(Cakes backed)"
df_profit.columns.name = "Demand(Cakes sold)"

df_profit
```

**Maximin(Wald) method**

```python
min_by_each_alt = df_profit.min(axis=1)

wald_best_value = min_by_each_alt.max()
wald_best_strategy = min_by_each_alt.idxmax()

df_styled = (
    df_profit.style.set_caption("Income Matrix (Wald-method)")
    .format(precision=0)
    .highlight_min(axis=1, color="#02290C")
)

display(df_styled)
print(
    f"Best alternative by Wald-method: {wald_best_strategy} (guaranteed income: {wald_best_value})"
)
```

**Minimax(Savage) method**

```python
# Find highest profit for each column
max_by_column = profit_matrix.max(axis=0)

# Subtract profits from maximums
risk_matrix = max_by_column - profit_matrix
df_risk = pd.DataFrame(
    risk_matrix,
    index=alternatives,
    columns=demand_values,
)
df_risk.index.name = "Alt-s risk"
df_risk.columns.name = "Demand(Cakes sold)"

# Find maximums from alt-s risks
max_risk_per_alt = df_risk.max(axis=1)

# Get minimum from the most profitable alt-s
minimax_best_value = max_risk_per_alt.min()
minimax_best_strategy = max_risk_per_alt.idxmin()

df_styled = (
    df_risk.style.set_caption("Risk Matrix (Savage Method)")
    .format(precision=0)
    .highlight_max(axis=1, color="#02290C")
)

# Display in Jupyter
display(df_styled)
print(
    f"Best alternative by Savage-method: {minimax_best_strategy}"
)
print(f"Minimax risk: {minimax_best_value}")
```

Bayes-Laplace approach (Rule of succession)

```python
sell_probabilities = days_count / days_count.sum()

alts_mean = df_profit.dot(sell_probabilities)
display(alts_mean)

bl_criteria = alts_mean.max()
bl_strategy = alts_mean.idxmax()

print(f"Best alternative by Bayes-Laplace approach: {bl_strategy}")
print(f"Best alternative expected income: {bl_criteria}")
```
